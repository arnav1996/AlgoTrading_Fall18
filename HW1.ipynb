```python
def DrawDown(Returns_Series):
    DD_list = []
    max1 = Returns_Series[0]
    min1 = Returns_Series[0]
    DD_value = (max1 - min1)/max1
    i = 1
    #index = 1
    while i<Returns_Series.size:
        if Returns_Series[i] > max1:
            if DD_value != 0:
                DD_list.append(DD_value)
                i = index + 1
                
            else:
                i += 1
                
            if i>=Returns_Series.size:
                break
                
            max1 = Returns_Series[i]
            min1 = Returns_Series[i]

            DD_value = (max1 - min1)/max1
            continue
            
        else:
            if Returns_Series[i] < min1:
                min1 = Returns_Series[i]
                index = i
                
            DD_value = (max1 - min1)/max1
            i += 1
        
        if i == Returns_Series.size - 1:
            if DD_value != 0:
                DD_list.append(DD_value)
            i = index + 1
            if i >= Returns_Series.size:
                break
                
            max1 = Returns_Series[i]
            min1 = Returns_Series[i]
            DD_value = (max1 - min1)/max1
            index = i
            continue
            
    return np.array(DD_list)
```

```python
def DD_largest_five(series_number):
    dd5_list = []
    for i in range(series_number.size):
        timeseries = cum_r[series_number[i]].values + 1
        dd5_list.append(np.sort(DrawDown(timeseries))[::-1][:5])
    return np.array(dd5_list)
```

```python
def calculate_winners_losers(ser):
    p_winners = []
    p_losers = []
    for i in ser:
        p = DF_five_series[DF_five_series>0][i].count()
        n = DF_five_series[DF_five_series<0][i].count()
        z = DF_five_series[DF_five_series==0][i].count()
        t = p+n+z
        p_winners.append(p*100/t)
        p_losers.append(n*100/t)
        
    return p_winners, p_losers
```
